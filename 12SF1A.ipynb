# Step 2 

### Initialized a new table called step2Table, indexed by Sample ID and Analyte name. The following code converts the columns into necessary datatype, and calculating the standard deviation while appending it to the table


```python
step2Table = pd.DataFrame({"Sample ID"   : rawData['Sample ID'],
                           "AnalyteName" : rawData['Analyte Name'],
                           "Conc (Calib)1": rawData['Conc (Calib)1'],
                           "Conc (Calib)2": rawData['Conc (Calib)2'],
                           "Conc (Calib)3": rawData['Conc (Calib)3'],
                          })
step2Table
step2Table = step2Table.fillna(value = 0.0)
step2Table = step2Table.replace(" ", 0.0)
step2Table = step2Table.set_index(["Sample ID", "AnalyteName"])
#May cause some sort of Roundoff Error
step2Table["Conc (Calib)1"] = step2Table["Conc (Calib)1"].astype(float)
step2Table = step2Table.astype({"Conc (Calib)1": 'float',
                                "Conc (Calib)2": 'float',
                                "Conc (Calib)3": 'float',
})
step2Table['sum'] = step2Table.sum(axis = 1)
step2Table = step2Table[step2Table['sum'] != 0]
step2Table = step2Table.drop(columns = 'sum')
std = step2Table.std(axis = 1)
avg = step2Table.mean(axis = 1)

step2Table['Average Conc Calib (mg/l)'] = avg
step2Table['Std Dev (mg/l)'] = std 
step2Table.tail(10)
#Step 2 Table is convered into format of 
#IndexType: Tuple ([Sample ID], [Analyte Name])
#All other columns are in float64 
```

### Step 3 Rearrange data so Sample and Analyte are on the right 

```python
step3Table = step2Table.drop(columns = ['Conc (Calib)1', 'Conc (Calib)2', 'Conc (Calib)3', ])
#10L1[01]01A20221129 --> 10B1[01]01A20221108
#Step3Table is a more precise table that only shows average concentration as well as standard deviation
step3Table
```

```python
def construct(table):
    #Implement SD next after checking accuracy
    ans = []
    avg = []
    std = []
    for info_tuple in table.index:
        b = info_tuple[0][2]
        if b != 'B' and len(info_tuple[0]) == 19:
            ans.append(info_tuple)
    for tuples in ans:
        if tuples in step3Table.index:
            avg.append(step3Table.loc[tuples]['Average Conc Calib (mg/l)'][0])
            std.append(step3Table.loc[tuples]['Std Dev (mg/l)'][0])
            
    ans_table = pd.DataFrame({
        "Samples" : ans,
        "Average Conc Calib (mg/l)": avg,
        "Average Std Dev (mg/l)": std
    })
    return ans_table

c_table = construct(step3Table)
#reset index to grab Blank
blank3reset = step3Table.reset_index()
c_table
```

```python
def calculate(constructed_table):
    ans = []
    for samples in constructed_table['Samples']:
        #Implementataion for now, SUPER SLOW TO BE FIXED LATER
        blank = ''
        for dilutions in dilutionTable['SA ID ']:
            if dilutions[0:4] == samples[0][0:2] + "B" + samples[0][3]:
                blank = (dilutions, samples[1])
                digestionVolumn = dilutionTable[dilutionTable['SA ID '] == samples[0]]['volume digestion (ml)'].unique().astype(float)[0]
                mass = dilutionTable[dilutionTable['SA ID '] == samples[0]]["mass sample (g)"].unique().astype(float)[0]
                dilution_constant = (dilutionTable[dilutionTable['SA ID '] == samples[0]]['volume digestion (ml)'].unique().astype(float)[0] / dilutionTable[dilutionTable['SA ID '] == samples[0]]["mass sample (g)"].unique().astype(float)[0])*    dilutionTable[dilutionTable['SA ID '] == samples[0]]["dilution factor"].unique().astype(float)[0]                                                 
                concInDigestate = constructed_table[constructed_table['Samples'] == samples]['Average Conc Calib (mg/l)'].unique().astype(float)[0]
                blankTbale = (blank3reset[blank3reset['Sample ID'] == blank[0]])
                blankValue = blankTbale[blankTbale["AnalyteName"] == blank[1]]["Average Conc Calib (mg/l)"].unique().astype(float)[0]
                if blankValue >= 0:
                    ans.append((concInDigestate - blankValue)*dilution_constant)
                else:
                    ans.append((concInDigestate)*dilution_constant)
    return ans 



c_table['Concentration of Element(ppm)'] = calculate(c_table)
c_table

```
